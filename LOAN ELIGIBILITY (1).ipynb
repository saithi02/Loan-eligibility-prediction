# Handling missing values in dataframes


```python
# Imputing null for categorical variables with mode
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Education'] = df['Education'].fillna(df['Education'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
# Imputing null for numerical variables with median - mean was not chosen because of outliers in the data
df['LoanAmount'].fillna(df['LoanAmount'].mean(), inplace = True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode(), inplace = True)
```

```python
df.LoanAmount_log=df['LoanAmount_log'].fillna(df.LoanAmount_log.mean())
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train = ss.fit_transform(X_train)
x_test = ss.fit_transform(X_test)
```

# Developing Classification Models

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf=RandomForestClassifier()
rf_clf.fit(X_train,y_train)
```

```python
from sklearn import metrics
y_pred=rf_clf.predict(x_test)
print("the accuracy of random forest classifier is",metrics.accuracy_score(y_pred,y_test)*100,"%")
y_pred
```

```python
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score, precision_score, recall_score, roc_auc_score, classification_report
```

```python
def draw_cm(actual,predicted):
    cm = confusion_matrix(actual,predicted)
    sns.heatmap(cm,annot=True, fmt='.2f', xticklabels=[0,1], yticklabels=[0,1])
    plt.ylabel('observed')
    plt.xlabel('Predicted')
    plt.show()
```

```python
draw_cm(y_test, y_pred)
```
